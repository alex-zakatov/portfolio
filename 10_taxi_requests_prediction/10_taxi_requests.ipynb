#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

### Вспомогательная функция

Разработаем функцию для быстрого расчёта и отображения тренда, сезонности и остатка.

* На входе — датафрейм
* На выходе:
    * Графики трёх компонент временного ряда.
    * Объект `seasonal_decompose`

```python
def decompose_and_plot(df):
    decomposed = seasonal_decompose(df)

    plt.figure(figsize=(16, 27))

    # Тренд
    plt.subplot(311)
    decomposed.trend.plot()
    plt.title('Тренд')
    plt.xlabel('Время')
    plt.ylabel('Количество заказов')

    # Сезонность
    plt.subplot(312)
    decomposed.seasonal.plot()
    plt.title('Сезонность')
    plt.xlabel('Время')
    plt.ylabel('Количество заказов')

    # Остаток
    plt.subplot(313)
    decomposed.resid.plot()
    plt.title('Остаток')
    plt.xlabel('Время')
    plt.ylabel('Количество заказов')

    plt.tight_layout()
    
    return decomposed
```

### Создание признаков

Отделим датафрейм, который будем использовать для машинного обучения:

```python
ml_df = df
```

Создадим дополнительные признаки:

#### Сalendar features
Так как в данных заметны недельная и суточная сезонности, полезными для обучения признаками могут стать час и день недели. Добавим их:

```python
ml_df['hour'] = ml_df.index.hour
ml_df['dayofweek'] = ml_df.index.day_of_week
```

#### Lag features
Добавим «отстающие значения» с зедержкой от 1 до 24 часов:

```python
for lag in list(range(1, 25)) + [169]:
        ml_df['lag_{}'.format(lag)] = ml_df['num_orders'].shift(lag)
        
# Удаление появившихся пропусков
ml_df = ml_df.dropna()
```

#### Rolling mean
Дополним датафрейм скользаящим средним за неделю (168 часов):

```python
ml_df['rolling_mean'] = ml_df['num_orders'].shift().rolling(168).mean()

# Удаление появившихся пропусков
ml_df = ml_df.dropna()

# Проверка
ml_df.head(3)
```

Дополнительные принаки созданы и добавлены в датафрейм. Образовавшиеся пропуски удалены.

### Разделение на выборки

Выделим обучающие признаки и целевую переменную. Параметр `shuffle = False` поможет сохранить исходную последовательность значений временного ряда.

Отделим тестовую выборку, размер которой по заданию — 10%. Валидацинная выборка будет составлять 10 от оставшихся 90%, предназначенных для обучения.

```python
train, test = train_test_split(ml_df, shuffle=False, test_size=0.1)

# Обучающая выборка
X_train = train.drop(['num_orders'], axis=1)
y_train = train['num_orders']

# Тестовая выборка
X_test = test.drop(['num_orders'], axis=1)
y_test = test['num_orders']

# Проверка
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

### Подготовка вспомогательных функций

Расчёт и отображение RMSE:

```python
# Среднее значение RMSE для кросс-валидации
def get_mean_cv_rmse(model, X, y):
   
    mean_RMSE = abs(cross_val_score(model, X, y,
                                    cv=TSCV,
                                    scoring='neg_root_mean_squared_error').mean())
    print(f"Среднее значение RMSE: {mean_RMSE}")
    return mean_RMSE

# RMSE для проверки на валидационной выборке
def get_rmse(y_true, y_pred):
    
    RMSE = round(mean_squared_error(y_true, y_pred)**0.5, 3)
    print(f"RMSE модели: {RMSE}")

    return RMSE
```

Расчёт и отображение времени:

```python
# Время обучения
def get_train_time (start):
    
    train_time = round(time.time() - start, 3)
    print(f"\nВремя обучения: {train_time} секунды")
    
    return train_time

# Время предсказания
def get_pred_time(start):
    
    pred_time = round(time.time() - start, 3)
    print(f"\nВремя предсказания: {pred_time} секунды" )
    
    return pred_time
```

## Обучение

Проанализируем работу трёх моделей:
* **Линеёная регрессия**
* **Случайный лес**
* **CatBoost**

Подберём гиперпараметры с помощью `GridSearchCV`, измерим время обучения и найдём среднюю метрику RMSE на кросс-валидации:

### Линейная регрессия

```python
# Инициализация
LR_model = LinearRegression()

# Подбор гиперпараметров
param_grid = {'fit_intercept': [True, False],
              'normalize': [True, False],
              'copy_X': [True, False],
              'n_jobs': [1, -1]}

# Инициализация и обучение GridSearch
grid_search = GridSearchCV(estimator=LR_model,
                           param_grid=param_grid,
                           cv=TSCV,
                           scoring='neg_root_mean_squared_error')
grid_search.fit(X_train, y_train)
print(f"Лушчие гиперпараметры:{grid_search.best_params_}")
```

```python
# Обучение оптимальной модели
start = time.time()
LR_model = LinearRegression(**grid_search.best_params_)
LR_model.fit(X_train, y_train)
LR_train_time = get_train_time(start)

LR_RMSE = get_mean_cv_rmse(LR_model, X_train, y_train)
```

### Случайный лес

```python
# Инициализация
RF_model = RandomForestRegressor()

# Сетка гиперпараметров
param_grid = {
    'n_estimators': range(1, 10),
    'max_depth': range(2, 5)
}

# Инициализация и обучение GridSearch
grid_search = GridSearchCV(estimator=RF_model,
                           param_grid=param_grid,
                           cv=TSCV,
                           scoring='neg_root_mean_squared_error')
grid_search.fit(X_train, y_train)

# Вывод наилучших гиперпараметров
print(f"Лушчие гиперпараметры:{grid_search.best_params_}")
```

```python
# Обучение оптимальной модели
start = time.time()
RF_model = RandomForestRegressor(**grid_search.best_params_,
                                 random_state = RANDOM_STATE)
RF_model.fit(X_train, y_train)
RF_train_time = get_train_time(start)

RF_RMSE = get_mean_cv_rmse(RF_model, X_train, y_train)
```

### CatBoost

```python
# Инициализация
CB_model = CatBoostRegressor(iterations=500,
                             loss_function='RMSE',
                             verbose=False,
                             random_state=RANDOM_STATE)
# Сетка параметров
param_grid = {
    'learning_rate': [0.1, 0.05],
    'max_depth': [3, 5, 7]
}

grid_search = GridSearchCV(CB_model,
                           param_grid,
                           cv=TSCV)
grid_search.fit(X_train, y_train)

# Вывод наилучших гиперпараметров
print(f"Лушчие гиперпараметры:{grid_search.best_params_}")
```

```python
# Обучение оптимальной модели
start = time.time()
CB_model = CatBoostRegressor(**grid_search.best_params_,
                             iterations=500,
                             loss_function='RMSE',
                             verbose=False,
                             random_state = RANDOM_STATE)
CB_model.fit(X_train, y_train)
CB_train_time = get_train_time(start)

CB_RMSE = get_mean_cv_rmse(CB_model, X_train, y_train)
```

### Сводные характеристики моделей
