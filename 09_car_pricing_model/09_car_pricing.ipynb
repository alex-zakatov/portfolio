## Обучение моделей

Подготовим вспомогательные функции для расчёта и отображения времени и RMSE моделей, чтобы уменьшить количество повторяющегося кода на этапе обучения и валидации. 

```python
# Время обучения
def get_train_time (start):
    train_time = round(time.time() - start, 3)
    print(f"\nВремя обучения: {train_time} секунды")
    return train_time

# Время предсказания
def get_pred_time(start):
    pred_time = round(time.time() - start, 3)
    print(f"\nВремя предсказания: {pred_time} секунды" )
    return pred_time

# RMSE
def get_rmse(y_true, y_pred):
    RMSE = round(mean_squared_error(y_true, y_pred)**0.5, 4)
    print(f"RMSE модели: {RMSE}")
    return RMSE
```

### LightGBM 

Выберем параметры и подготовим данные:

```python
params = {
    'boosting_type': 'gbdt',    # Алгоритм бустинга
    'objective': 'regression',  # Цель
    'metric': 'rmse',           # Метрика (по заданию RMSE)
    'num_leaves': 20,           # Количество листьев дерева (по умолчанию 31)
    'learning_rate': 0.05,      # Скорость обучения (по умолчанию 0.1)
    'feature_fraction': 0.9,    # Выбор доли показателей из общего числа для обучения на каждой итерации
    'bagging_fraction': 0.8,    # Часть данных, которая будет случайно отбираться при обучении
    'bagging_freq': 5,          # Частота отбора
    'verbose': 0,               # Управление логирований сообщений при обучении
    'n_jobs': -1,               # Распараллеливание ('-1' — все доступные ядра )
    'force_col_wise': True,     # Построение гистограммы по столбцам (True — повышение скорости)
    'feature_pre_filter': False # Масштабирование (False — отключено)
}

# Подготовка обучающей и валидационной выборок
lgbm_train = lgbm.Dataset(X_train, y_train)
lgbm_val = lgbm.Dataset(X_val, y_val)
```

Обучим модель и отобразим время обучения:

```python
# Начало отсчёта времени
start = time.time()

# Тренировка модели
LGBM_model = lgbm.train(params,
                        lgbm_train,
                        num_boost_round=500,
                        valid_sets=[lgbm_train, lgbm_val],
#                         early_stopping_rounds=60,
#                         verbose_eval=100
                       )
# Конец отсчёта
LGBM_train_time = get_train_time(start)
```

Найдём время, за которое модель выполнит предсказание, и проверим метрику RMSE:

```python
start = time.time()

# Проверка на валидационной выборке
y_pred = LGBM_model.predict(X_val)

# Время и метрики
LGBM_pred_time = get_pred_time(start)
LGBM_RMSE = get_rmse(y_val, y_pred)
```
